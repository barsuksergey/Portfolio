```python
%matplotlib notebook
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

revenue = [89, 109, 110, 130, 119, 127, 128, 96, 134, 135, 137, 141, 171, 183]
month_cpi_2020 = [257.971, 258.678, 258.115, 256.389, 256.394, 257.797]
cpi2020 = np.mean(month_cpi_2020)
cpi = [201.6, 207.3, 215.303, 214.537, 218.056, 224.939, 229.594, 232.957, 236.736, 237.017, 240.007, 245.120, 251.107, 255.657]

GP = [82, 82, 82, 82, 82, 82, 82, 48, 82, 82, 82, 82, 82, 82]
W = [58, 50, 54, 51, 44, 47, 48, 24, 39, 43, 41, 33, 30, 32]

W_percent = []
for x in range(14):
    W_percent.append(W[x]/GP[x])

adj_revenue = []
for x in range(14):
    adj_revenue.append(revenue[x]*(cpi2020/cpi[x]))

season = []
for x in range(2005,2019):
    season.append('{}/{}'.format(x, x+1))
    x += 1
    
data = {'win' : W_percent, 'revenue' : adj_revenue, 'season' : season}
df = pd.DataFrame(data=data)
```

```python
sns.regplot(x='win', y='revenue', data = df, color = 'red', marker = 'x')
plt.ylabel('Revenue, millions $')
plt.xlabel('Win rate, %')
plt.title('Detroit Red Wings revenue to win rate')

plt.savefig('Assignment 4', dpi = 400)
```
